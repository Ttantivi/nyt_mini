```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.ExcelFile('data/NYT_mini_times.xlsx')
df = data.parse('Sheet1')
df['Date'] = pd.to_datetime(df['Date']).dt.date
```

```python
# Define a figure
plt.figure(figsize=(10,7))

# Getting column names
columns = df.columns
columns = columns[1:] # don't care about date column
colors = ['blue', 'red', 'green', 'cyan', 'black', 'yellow', 'pink']
 
# Plot the KDE for each column
for column, color in zip(columns, colors):
    sns.kdeplot(df[column], color=color, alpha=0.5, fill=True, label=column)

# Set a legend
plt.legend(title = "Peoples")
plt.xlabel('Seconds')

# Show the plot
plt.show()
```

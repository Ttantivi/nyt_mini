```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.ExcelFile('data/NYT_mini_times.xlsx')
df = data.parse('Sheet1')
df['Date'] = pd.to_datetime(df['Date']).dt.date
```

```python
# Define a figure
plt.figure(figsize=(10,7))

# Getting column names
columns = df.columns
columns = columns[1:] # don't care about date column
colors = ['blue', 'red', 'green', 'cyan', 'white', 'yellow', 'pink']

# Melt the dataframe to long format
df_melted = pd.melt(df, value_vars=columns, var_name='Peoples', value_name='Seconds')

# Define a figure
plt.figure(figsize=(10,7))

# Create a boxplot
sns.boxplot(x='Peoples', y='Seconds', data=df_melted, palette=colors)

# Show the plot
plt.show()
```
